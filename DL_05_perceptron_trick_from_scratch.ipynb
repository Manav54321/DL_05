```python
def perceptron(X, y):
  X = np.insert(X, 0, 1, axis = 1) # insert one in first column of the X array for bias
  weights = np.ones(X.shape[1]) # initial weights
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0, X.shape[0]) # for selecting a random number
    y_hat = step(np.dot(X[j], weights))
    weights = weights + lr * (y[j] - y_hat) * X[j]

  return weights[1:], weights[0]
```

```python
def step(z):
  return 1 if z > 0 else 0
```

```python
step(0.294)
```

```python
step(-0.294)
```

```python
coef_, intercept_ = perceptron(X, y)
```

```python
print(intercept_)
print(coef_)
```

```python
m = -coef_[0]/coef_[1]
b = -intercept_/coef_[1]
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
